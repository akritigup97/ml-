```python
import pandas as pd 
ratings= pd.read_csv("/Users/akriti/Desktop/development/Datasets/movie_collab/ratings.csv")
movies= pd.read_csv("/Users/akriti/Desktop/development/Datasets/movie_collab/movies.csv")
data= pd.merge(movies,ratings).drop(['genres','timestamp'],axis=1)
data.head()
```

```python
user_ratings= data.pivot_table(index=['userId'],columns=['title'],values='rating')
user_ratings.head()
```

```python
user_ratings= user_ratings.dropna(thresh=100,axis=1).fillna(0)# axis is 1 as we are working on columns

```

```python

item_similarity_df = user_ratings.corr(method='pearson')
item_similarity_df.head()
```

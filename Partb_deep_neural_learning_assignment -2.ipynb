```python
import numpy as np
import pandas as pd
from statistics import pstdev
from statistics import mean 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

the below line of code is used to load the data 


```python
dataset=pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
#to display the first 5 rows of a dataset we use head 
dataset.head()
```

The below code is used to separate the features or the factors and the label column i.e. strength 

```python
n_cols= dataset.columns
features = dataset[n_cols[n_cols!='Strength']]
target= dataset['Strength']

```

Below we are viewing the features and labels separately 

```python
features.head()
```

```python
target.head()
```

```python
features_norm = (features - features.mean()) / features.std()
features_norm.head()
```

```python
n_input=features_norm.shape[1]
```

In few imports below we are fetching the sequential model  and dense layer for building our model 

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense 
```

the function below is used to create the regression model 

```python
def regression_model():
    model= Sequential()
    model.add(Dense(10,activation="relu",input_shape=(n_input,)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
   
    return model
```

Now we make a call to the regression model 

```python
model = regression_model()
```

The dataset is split between training and testing . 
Error is a list used to store all the mean squared errors 


```python
error =[]
for i in range(50):
    X_train,X_test,y_train,y_test= train_test_split(features,target,random_state=0,test_size=0.3)
    #training the model 
    model.fit(X_train, y_train,epochs=50, verbose=2)
    #predicting values for unknown test data 
    y_predicts = model.predict(X_test)
    #calculating mean_square_error
    mean_err = mean_squared_error(y_test, y_predicts)
    #adding all the mean_err to error list
    error.append(mean_err)
```

```python
#calculating the mean of errors 
mean(error)

```

```python
#calculating the standard deviation
pstdev(error)
```
